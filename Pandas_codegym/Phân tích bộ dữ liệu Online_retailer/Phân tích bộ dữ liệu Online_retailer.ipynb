1. Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;<br>

```python
data['Month'] = pd.to_datetime(data['Date']).dt.month
data['Month'].info()
condition = [(data['Month'] ==1) | (data['Month'] ==2) | (data['Month'] ==3), 
             (data['Month'] ==4) | (data['Month'] ==5) | (data['Month'] ==6),
             (data['Month'] ==7) | (data['Month'] ==8) | (data['Month'] ==9),
             (data['Month'] ==10) | (data['Month'] ==11) | (data['Month'] ==12)]
choices =[1,2,3,4]
data['Previous'] = np.select(condition, choices)
data.head(5)
```

3. Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.<br>


```python
condition_discount = [(data['Country']=='United Kingdom') & (data['Previous']==4), (data['Country']=='France')]
choice_discount =[0.1, 0.05]
data['Discount'] = np.select(condition_discount, choice_discount, 0)
data.head(5)
```

4. Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount

```python
data['Total Revenues after Discount'] = data['Revenues'] - data['Revenues']*data['Discount']
```
